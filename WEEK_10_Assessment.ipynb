Solving Unknown variables using Cramer's Rule


```python
import numpy as np
A = np.array([[1,1,1],[1,2,2],[1,2,3]])
print(A) 
print("|A| =", + round(np.linalg.det(A))) #determinant of the original matrix
print("----------")
A1 = np.array([[5,1,1],[6,2,2],[8,2,3]])
print(A1)  
print("|A1| =", + round(np.linalg.det(A1))) #determinant of the matrix A1
print("----------")
A2 = np.array([[1,5,1],[1,6,2],[1,8,3]])
print(A2)
print("|A2| =", + round(np.linalg.det(A2))) #determinant of the matrix A2
print("----------")
A3  = np.array([[1,1,5],[1,2,6],[1,2,8]])
print(A3)
print("|A3| =", + round(np.linalg.det(A3))) #determinant of the matrix A3
print("----------")
X1 = round(np.linalg.det(A1)) / round(np.linalg.det(A))
print("X1 =", + X1)
X2 = round(np.linalg.det(A2)) / round(np.linalg.det(A))
print("X2 =", + X2)
X3 = round(np.linalg.det(A3)) / round(np.linalg.det(A))
print("X3 =", + X3)
```
